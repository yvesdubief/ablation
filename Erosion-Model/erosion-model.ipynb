## Level-set method

$$
\frac{\partial \phi}{\partial t} = (\alpha \kappa-u)\left\vert\boldsymbol{\nabla}\phi\right\vert
$$

```python
""" Initialization: loading libraries, setting up parameters"""
import matplotlib.pyplot as plt
import skfmm 
import numpy as np
from scipy.interpolate import interp1d
from scipy.interpolate import RegularGridInterpolator
from scipy.interpolate import griddata
L = 2.
N = 51
R = 1
delta = 2*L/(N-1)
x = np.linspace(-L,L,N)
y = np.linspace(-L,L,N)
X,Y = np.meshgrid(x,y)

# loading wall shear stress data
data = np.genfromtxt('shear-stress.txt')
# create interpolation function allowing for extrapolation between ends
fwallshear = interp1d(data[:,0], data[:,1],bounds_error=False, fill_value=0.0645)
```

The level set method requires to project the arclength data on the grid. We use python function griddata, however to be successful, other points must be added throughout the domain. This first version uses random points.

```python
nrndpts = 5000
rndpts = np.zeros((nrndpts,2))
erosionrndpts = np.zeros(nrndpts)
phirndpts = np.zeros(nrndpts)
dphirndpts = np.zeros((nrndpts,2))
rndpts[:,0] = np.random.uniform(L,-L,nrndpts)
rndpts[:,1] = np.random.uniform(L,-L,nrndpts)
```

Time advancement uses a first order Euler method. At each time step, the contour $\phi=0$ is extracted and seeded with points using matplotlib functions. Randomly seeded points which are too close to the solid surface are relocated on the domain edges. Efficiency and accuracy must be improved in future version.

```python
Nt = 200
dt = 0.01
Ntreinit = 10
cst = -1.
# initial shape
phi = X**2 + Y**2 - R
phi = skfmm.distance(phi, dx = delta )
phio = np.copy(phi)
it = 0
fig = plt.figure(num=None, figsize=(4, 4), dpi=160, facecolor='w', edgecolor='k')
ax = fig.add_subplot(111)
while it< Nt:
    # compute gradient of phi and its norm
    dphi = np.array(np.gradient(phi,delta,delta))
    normdphi = np.sqrt(dphi[0,:,:]**2 + dphi[1,:,:]**2)
    # extract phi=0 contour
    CS = ax.contour(X,Y,phi, levels = [0])
    # seed contour with points
    dat0= CS.allsegs[0][0]
    # find leading edge and rearrange point data
    istart = np.argmin(np.abs(dat0[:,0] + 1 )+np.abs(dat0[:,1]))
    n = len(dat0[:,0])
#     pcolor= np.zeros(n) # used to visually verify that new points are set right
#     newpcolor= np.zeros(n)
    segment = np.zeros(n)
    arclength = np.zeros(n)
    newdat0 = np.zeros((n,2))
    icolor = 0
    for i in range(istart,n):
#         pcolor[i] = icolor
#         newpcolor[icolor] = icolor
        newdat0[icolor,:] = dat0[i,:]
        icolor += 1
    for i in range(0,istart):
#         pcolor[i] = icolor
#         newpcolor[icolor] = icolor
        newdat0[icolor,:] = dat0[i,:]
        icolor += 1
    # find randomly placed points of concern and move them away from boundary using local gradient
    fphi = RegularGridInterpolator((x,y),phi)
    fdphi0 = RegularGridInterpolator((x,y),dphi[0,:,:])
    fdphi1 = RegularGridInterpolator((x,y),dphi[1,:,:])
    phirndpts = fphi(rndpts)
    dphirndpts[:,0] = fdphi0(rndpts)
    dphirndpts[:,1] = fdphi1(rndpts)
    maskrndpts = np.where(phirndpts <= delta)
    rndpts[maskrndpts][:] += dphirndpts[maskrndpts][:]
#     for i in range(nrndpts):
#         for j in range(n):
#             r2 = (newdat0[j,0]-rndpts[i,0])**2+(newdat0[j,1]-rndpts[i,1])**2
#             if np.sqrt(r2) < delta:
#                 rndpts[i,:] = 0
                
        
    segment[0] = np.sqrt((newdat0[0,0] + 1 )**2+(newdat0[0,1])**2)
    segment[1:] = np.sqrt((newdat0[1:,0] - newdat0[0:-1,0])**2
                     +(newdat0[1:,1]- newdat0[0:-1,1])**2)
#    plt.scatter(newdat0[:,0],newdat0[:,1], newpcolor)
    
    arclength = np.cumsum(segment)
    erosion = fwallshear(arclength/arclength[-1])
    newnewdat0 = np.concatenate((newdat0,rndpts),axis=0)
    newerosion = np.concatenate((erosion,erosionrndpts), axis = 0)
    speed = griddata(newnewdat0, newerosion, (X, Y), method='cubic')
    np.nan_to_num(speed,False)
#    print(it,speed.shape)
##    dphi = np.array(np.gradient(phi,delta,delta))
##    udphi = speed[0,:,:]*dphi[0,:,:]+speed[1,:,:]*dphi[1,:,:]
##    mask = np.where(udphi > 0)
##    udphi[mask] = 0.
    phi -= dt*cst*speed*normdphi
#    if (it%Ntreinit == 0):
#        phi = skfmm.distance(phi, dx = delta )
    it += 1
plt.show()
```

```python
fig = plt.figure(num=None, figsize=(4, 4), dpi=160, facecolor='w', edgecolor='k')
ax = fig.add_subplot(111)
# im = ax.contour(X,Y,phi,21, cmap = 'jet')
CS = ax.contour(X,Y,phio, levels = [0])
levs = np.linspace(-1,1,51)
ax.contourf(X,Y,speed, levels = levs, cmap = 'jet')
ax.contour(X,Y,phi, levels = [0])
plt.scatter(newnewdat0[:,0],newnewdat0[:,1],(newerosion*100+1))
#dat0= CS.allsegs[0][0]
ax.set_aspect('equal')
plt.show()
```

Random seeding cannot only be improved by extending the erosion rate to points in close vicinity to the surface. In any case, this failed version, shows that randomization introduces too much spatial errors, most likely from back and forth interpolation.

In the second version, particles are seeded uniformily on both sides of the interface following the local normal.

Surface tension is added through the curvature of the solid/fluid interface:
$$
\kappa = \frac{1}{2}\nabla\cdot\frac{\nabla\phi}{\left\vert\nabla\phi\right\vert}
$$
and governed by the parameter $\alpha$.

```python
def curvature(phi,dphi):
    global N,delta
    delta2 = delta**2
    dxx = np.zeros((N,N))
    dyy = np.zeros((N,N))
    dxy = np.zeros((N,N))
    dxx[1:-1,:] = (phi[:-2,:] - 2*phi[1:-1,:] + phi[2:,:])/delta2
    dxx = neumann(dxx,0)
    dyy[:,1:-1] = (phi[:,:-2] - 2*phi[:,1:-1] + phi[:,2:])/delta2
    dyy = neumann(dyy,0)
    dxy[1:-1,1:-1] = (phi[2:,2:] - phi[2:,:-2] + phi[:-2,:-2] - phi[:-2,2:])/(4*delta2)
    dxy = neumann(dxy,0)
    dxy = neumann(dxy,1)
    return (dphi[0,:,:]**2*dyy + dphi[1,:,:]**2*dxx - 2*dphi[0,:,:]*dphi[1,:,:]*dxy)/ \
            (2*(dphi[0,:,:]**2 + dphi[1,:,:]**2))
def neumann(d,iaxis):
    if iaxis == 0:
        d[0,:] = d[1,:]
        d[-1,:] = d[-2,:]
    else:
        d[:,0] = d[:,1]
        d[:,-1] = d[:,-2]
    return d
```

```python
Nt = 300
dt = 0.001
Ntreinit = 5
cst = -1.
alpha = 0.01
nextras_per_seg = 5
# initial shape
phi = X**2 + Y**2 - R
phi = skfmm.distance(phi, dx = delta )
phio = np.copy(phi)
it = 0
fig = plt.figure(num=None, figsize=(4, 4), dpi=160, facecolor='w', edgecolor='k')
ax = fig.add_subplot(111)
while it< Nt:
    # compute gradient of phi and its norm
    dphi = np.array(np.gradient(phi,delta,delta))
    normdphi = np.sqrt(dphi[0,:,:]**2 + dphi[1,:,:]**2)
    dphi[0,:,:] /= normdphi
    dphi[1,:,:] /= normdphi
    fphi = RegularGridInterpolator((x,y),phi)
    fdphi0 = RegularGridInterpolator((x,y),dphi[0,:,:])
    fdphi1 = RegularGridInterpolator((x,y),dphi[1,:,:])
    # extract phi=0 contour
    CS = ax.contour(X,Y,phi, levels = [0])
    # seed contour with points
    dat0= CS.allsegs[0][0]
    # find leading edge and rearrange point data
    istart = np.argmin(np.abs(dat0[:,0] + 1 )+np.abs(dat0[:,1]))
    n = len(dat0[:,0])
#     pcolor= np.zeros(n) # used to visually verify that new points are set right
#     newpcolor= np.zeros(n)
    nextras = n*nextras_per_seg*2
    extrapts = np.zeros((nextras,2))
    erosionextrapts = np.zeros(nextras)
    phiextrapts = np.zeros(nextras)
    dphiextrapts = np.zeros((nextras,2))

    segment = np.zeros(n)
    arclength = np.zeros(n)
    newdat0 = np.zeros((n,2))
    dphisegs = np.zeros((n,2))
    
    icolor = 0
    for i in range(istart,n):
#         pcolor[i] = icolor
#         newpcolor[icolor] = icolor
        newdat0[icolor,:] = dat0[i,:]
        icolor += 1
    for i in range(0,istart):
#         pcolor[i] = icolor
#         newpcolor[icolor] = icolor
        newdat0[icolor,:] = dat0[i,:]
        icolor += 1
    # find randomly placed points of concern and move them away from boundary using local gradient

#     dphirndpts[:,0] = fdphi0(rndpts)
#     dphirndpts[:,1] = fdphi1(rndpts)
#     maskrndpts = np.where(phirndpts < delta)
#     extrapts[maskrndpts][:] += dphirndpts[maskrndpts][:]
#     for i in range(nrndpts):
#         for j in range(n):
#             r2 = (newdat0[j,0]-rndpts[i,0])**2+(newdat0[j,1]-rndpts[i,1])**2
#             if np.sqrt(r2) < delta:
#                 rndpts[i,:] = 0
                
        
    segment[0] = np.sqrt((newdat0[0,0] + 1 )**2+(newdat0[0,1])**2)
    segment[1:] = np.sqrt((newdat0[1:,0] - newdat0[0:-1,0])**2
                     +(newdat0[1:,1]- newdat0[0:-1,1])**2)
#    plt.scatter(newdat0[:,0],newdat0[:,1], newpcolor)
    
    arclength = np.cumsum(segment)
    erosion = fwallshear(arclength/arclength[-1])
    
    dphisegs[:,0] = fdphi0(newdat0)
    dphisegs[:,1] = fdphi1(newdat0)
    iextra = 0
    for iseg in range(n):
        for iext in range(nextras_per_seg):
            extrapts[iextra,:] = newdat0[iseg,:] + (iext + 1) * delta * dphisegs[iseg,:]
            erosionextrapts[iextra] = erosion[iseg]
            iextra += 1
            extrapts[iextra,:] = newdat0[iseg,:] - (iext + 1) * delta * dphisegs[iseg,:]
            erosionextrapts[iextra] = erosion[iseg]
            iextra += 1
    
    phiextrapts = fphi(extrapts)
    newnewdat0 = np.concatenate((newdat0,extrapts),axis=0)
    newerosion = np.concatenate((erosion,erosionextrapts), axis = 0)
    speed = griddata(newnewdat0, newerosion, (X, Y), method='cubic')
    np.nan_to_num(speed,False)
#    print(it,speed.shape)
##    dphi = np.array(np.gradient(phi,delta,delta))
##    udphi = speed[0,:,:]*dphi[0,:,:]+speed[1,:,:]*dphi[1,:,:]
##    mask = np.where(udphi > 0)
##    udphi[mask] = 0.
    kappa = curvature(phi,dphi)
    phi += dt*(alpha*kappa - cst*speed)*normdphi
#    if (it%Ntreinit == 0):
#        phi = skfmm.distance(phi, dx = delta )
    it += 1
plt.show()
```
